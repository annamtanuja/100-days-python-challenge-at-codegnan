## Types of Arguments
1.positional Arguments
2.keyword Arguments
3.Default Arguments
4.variable lenghts positional of agrument
5.variable lenght of keyword arguments

```python
# default arguments  
def add(a,b):
    return a+b
print(add(10,2))
```

```python
# default arguments
def add(a,b,c=0,d=0):
    return a+b+c+d 
print(add(1,2))
print(add(1,2,3))
print(add(1,2,4))
```

```python
# multiplication of numbers using default arguments
def mul(a,b,c=1,d=1):
    return a*b*c*d 
print(mul(1,2,3,4))
print(mul(1,2,3))
print(mul(1,2,4))
print(mul(1,2))
```

```python
# variable length of positional arguments
def add(a,*args):
    print(a)
    print(args)
add(1,2,3,4,5,6)
```

```python
# variable length of positional arguments
a,*b,c=10,20,30,40,50
print(a)
print(b)
print(c)
```

```python
a,*b=10,20,30,40,50
print(a)
print(b)
```

```python
# variable length of positional arguments
def add(a,*args):
    print(a)
    print(args)
    s=0
    for num in args:
        s+=num
    return s
print(add(1,2,3,9))
```

```python
# convert list of numbers into even list and odd list
lst = list(map(int, input().split()))
even_list = []
odd_list = []
for num in lst:
    # check the numbers os even or odd
    if num % 2 == 0:
        even_list.append(num)
    else:
        odd_list.append(num)
print("Even list:",even_list)
print("Odd list:",odd_list)
```

```python
# marks analyser
#in list contains students marks, in that, find the pass percentage and fail percentage
#pass marks is 35(+)
#fail is below 35 (-)
marks=list(map(int, input("enter marks:").split()))
def marks_analyser(marks):
    avg = sum(marks)/len (marks)
    if avg >= 35:
        return "pass"
    else:
        return "fail"
print(marks_analyser(marks))

```
