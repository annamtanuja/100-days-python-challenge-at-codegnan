# Nested loops

```python
# A loop contain one more loop that we can call it as nested loop
for i in range (1,3): # outer block
  for j in range (1,4):  # inner block
    # block of code
     print(i,j)
```

```python
# print 1 to 10 numbers
for i in range(1,11):
    print(i)
```

```python
# print 1 to 10 numbers
i = 0
while i < 11:
    print(i)
```

```python
for i in range (1,3): # outer block
  for j in range (1,11):  # inner block
    # block of code
     print(f"{i} * {j} = {i*j}")
  print("_____________ ")
```

```python
# print i1 to 10 numbers 
i=0
for i in range (1,3):
    for j in range (1,11):
        print(f"{i} * {j} = {i*j}")
   
```

```python
# write 3 table program in functions
def print_table (n):
  for i in range (1,11):
    print(f"{n} * {i} = {n*i}")
  return   
# 3 table
print_table(3)

    
```
